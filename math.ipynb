# Linear Combination

### Linear Combination

A **linear combination** is just a fancy term for **mixing numbers together using multiplication and addition**. Specifically:
- You take a set of numbers (called **inputs** or **features**),
- Multiply each number by a corresponding weight (a scaling factor),
- Add up all the results to get a single output.

Mathematically:
$$
y = w_1x_1 + w_2x_2 + \dots + w_nx_n
$$
Where:
- $x_1, x_2, \dots, x_n$: Input values (e.g., pixel values, sensor readings, etc.),
- $w_1, w_2, \dots, w_n$: Weights (how important each input is),
- $y$: The final result (output).

### Simple Example

Imagine you‚Äôre making a fruit salad, and you want to calculate the total cost based on how many apples and bananas you buy:
- Apples cost $2 each,
- Bananas cost $1 each.

If you buy:
- 3 apples ($x_1 = 3$),
- 5 bananas ($x_2 = 5$),

The total cost is:
$$
\text{Total Cost} = (2 \cdot 3) + (1 \cdot 5) = 6 + 5 = 11
$$

Here:
- The weights are the prices ($w_1 = 2$, $w_2 = 1$),
- The inputs are the quantities ($x_1 = 3$, $x_2 = 5$),
- The result ($y = 11$) is the total cost.

### Why Is It Called "Linear"?

It‚Äôs called **linear** because:
- You‚Äôre only multiplying and adding‚Äîno exponents, square roots, or other fancy math.
- The relationship between the inputs and the output is a straight line (if graphed).

# Logarithm

A logarithm answers the question:
> "To what power must a base number be raised to get another number?"

In mathematical terms:
$$
\log_b(x) = y \quad \text{means} \quad b^y = x
$$

- $b$: The **base** (the number being raised to a power).
- $x$: The **result** of raising $b$ to the power $y$.
- $y$: The **logarithm** (the exponent you‚Äôre solving for).

### **Key Idea**
Logarithms are like the "opposite" of exponents:
- Exponentiation: $b^y = x$ (start with $b$ and $y$, find $x$).
- Logarithm: $\log_b(x) = y$ (start with $b$ and $x$, find $y$).

### **Examples**

**Example 1: Base 10**
$$
\log_{10}(100) = 2 \quad \text{because} \quad 10^2 = 100
$$

**Example 2: Base 2**
$$
\log_2(8) = 3 \quad \text{because} \quad 2^3 = 8
$$

**Example 3: Natural Logarithm ($\ln$)**
The natural logarithm uses base $e$ (approximately $2.718$):
$$
\ln(7.389) \approx 2 \quad \text{because} \quad e^2 \approx 7.389
$$

### **Why Are Logarithms Useful?**
1. **Simplify Large Numbers**:
   - Logarithms turn multiplication into addition, division into subtraction, and exponents into multiplication.
   - Example: $\log(ab) = \log(a) + \log(b)$.

2. **Measure Growth Rates**:
   - Used in fields like finance (e.g., compound interest), biology (e.g., population growth), and machine learning (e.g., loss functions).

3. **Scale Data**:
   - Logarithms compress large ranges of values into smaller ones (e.g., decibels for sound, pH for acidity).

### **Key Takeaways**
1. A logarithm tells you the **exponent** needed to raise a base to get a result.
2. Examples:
   - $\log_{10}(100) = 2$
   - $\log_2(8) = 3$
   - $\ln(7.389) \approx 2$
3. Logarithms simplify calculations, measure growth, and scale data.

# Element-Wise Multiplication

Great questions! Let‚Äôs break down **element-wise multiplication** and the **dot product** step by step, with examples to make it clear.

### 1. **Element-Wise Multiplication**

#### What Is It?
- **Element-wise multiplication** multiplies corresponding elements of two tensors (or vectors) of the same size.
- The result is a new tensor where each element is the product of the corresponding elements from the input tensors.

#### Example:
Let‚Äôs say we have two tensors:
```python
A = [1, 2, 3]
B = [4, 5, 6]
```

The element-wise multiplication is:
```python
C = [1*4, 2*5, 3*6] = [4, 10, 18]
```

#### When Is It Used?
- Element-wise multiplication is often used in operations like scaling or masking tensors.
- It‚Äôs **not typically used for computing predictions in linear models**, because it doesn‚Äôt sum up the results into a single value.

### 2. **Dot Product**

#### What Is It?
- The **dot product** is a specific operation that combines two vectors (or tensors) into a **single scalar value**.
- It involves:
  1. Multiplying corresponding elements of the two vectors.
  2. Summing up all the products.

Mathematically:
$$
\text{Dot Product} = \sum_{i=1}^n (A[i] \cdot B[i])
$$

#### Example:
Using the same tensors as above:
```python
A = [1, 2, 3]
B = [4, 5, 6]
```

The dot product is:
$$
\text{Dot Product} = (1 \cdot 4) + (2 \cdot 5) + (3 \cdot 6) = 4 + 10 + 18 = 32
$$

#### When Is It Used?
- The dot product is fundamental in machine learning and deep learning.
- It‚Äôs used to compute the **weighted sum** of inputs in linear models:
  $$
  y = w \cdot x + b
  $$
  Here:
  - $w$ (weights) and $x$ (inputs) are vectors,
  - Their dot product gives a single scalar value ($y$).

### 3. **Key Differences Between Element-Wise Multiplication and Dot Product**

| Feature                     | Element-Wise Multiplication          | Dot Product                         |
|-----------------------------|---------------------------------------|--------------------------------------|
| **Output Shape**            | Same shape as input tensors          | Single scalar value                 |
| **Operation**               | Multiplies corresponding elements    | Multiplies and sums corresponding elements |
| **Use Case**                | Scaling, masking, etc.               | Computing weighted sums in models   |

### 4. **Why Does This Matter in Machine Learning?**

In machine learning:
- **Element-wise multiplication** is less common for predictions but might be used in operations like applying masks or scaling activations.
- **Dot product** is crucial because it‚Äôs the basis of how models combine inputs and weights to make predictions:
  $$
  \text{Prediction} = \text{Input} \cdot \text{Weights} + \text{Bias}
  $$

For example:
- If `train_x[0]` is an input vector (e.g., flattened image) and `weights` is a weight vector, their dot product computes the weighted sum of the inputs.

Mathematically, sigmoid is represented as:<br>


$$ \sigma(x) = \frac{1}{1 + e^{-x}} $$


where, 
- $x$ is the input value,
- $e$ is Euler‚Äôs number $(‚âà2.718)$

Sigmoid function is used as an activation function in machine learning and neural networks for modeling binary classification problems, smoothing outputs, and introducing non-linearity into models.

### Scenario:
Imagine you're working with a dataset where you have two variables:
- `x` (input): The size of the house in square feet.
- `y` (output): The price of the house in thousands of dollars.

You want to create a simple linear model to predict the price (`y`) of a house given its size (`x`). A linear model can be represented as:

$$
y = mx + b
$$

Where:
- $m$ is the slope (how much the price changes per square foot).
- $b$ is the y-intercept (the predicted price when the house size is 0).

### Example Data:
Let's say you have the following data points:

| House Size (x, in sq ft) | Price (y, in $1000s) |
|---------------------------|-----------------------|
| 1000                      | 200                  |
| 1500                      | 300                  |

### Step 1: Calculate the Slope ($m$)
The formula for the slope between two points $(x_1, y_1)$ and $(x_2, y_2)$ is:

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

Using our data points:
- $(x_1, y_1) = (1000, 200)$
- $(x_2, y_2) = (1500, 300)$

$$
m = \frac{300 - 200}{1500 - 1000} = \frac{100}{500} = 0.2
$$

So, the slope $m = 0.2$. This means that for every additional square foot, the price increases by $200 (since $0.2 \times 1000 = 200$).

### Step 2: Calculate the Y-Intercept ($b$)
The y-intercept $b$ is the value of $y$ when $x = 0$. We can use the equation of the line $y = mx + b$ and one of the data points to solve for $b$.

Using the point $(1000, 200)$:
$$
200 = 0.2(1000) + b
$$
$$
200 = 200 + b
$$
$$
b = 0
$$

So, the y-intercept $b = 0$. This means that if the house size is 0 square feet, the predicted price would be $0.

### Final Linear Model:
Now we have the complete linear equation:
$$
y = 0.2x + 0
$$

Or simply:
$$
y = 0.2x
$$

### Real-World Prediction:
If you want to predict the price of a house that is 2000 square feet:
$$
y = 0.2(2000) = 400
$$

This means the predicted price of a 2000 square foot house is $400,000.

### Summary:
- **Slope ($m$)**: Represents the rate of change of price with respect to house size (in this case, $0.2$ or $200 per square foot).
- **Y-Intercept ($b$)**: Represents the base price when the house size is 0 (in this case, $0).

This is a very basic example of how slope and intercept are used in a real-world machine learning scenario to make predictions.

The name Quadratic comes from "quad" meaning square, because the variable gets squared (like x2). It has 3 variables which if you look at the curve, it represents beginning, center, ending of the curved line.

A quadratic equation is a type of mathematical equation where the highest power of the variable (usually x) is 2. In simpler terms, it's an equation that involves a squared term $(x2)$, and no higher powers like $x^3, x^4$ etc.

The general form of a quadratic equation is:

$$ax2+bx+c=0$$

Where:
a, b, and c are constants (numbers),
a != 0 (because if a=0, the equation would no longer have an $x^2$ term and wouldn't be quadratic).

But sometimes a quadratic equation does not look like that! There are more types of quadratic equtions. But there is always a,b and c.

In a quadratic equation, the variables and constants $a$, $x$, $b$, and $c$ each have specific roles. Let's break them down in detail:

#### 1. **$a$: The Coefficient of $x^2$**
- $a$ is the coefficient of the squared term ($x^2$).
- It determines the "shape" and "direction" of the parabola (the graph of the quadratic equation):
  - If $a > 0$, the parabola opens **upwards** (like a smile).
  - If $a < 0$, the parabola opens **downwards** (like a frown).
- $a$ also affects how "wide" or "narrow" the parabola is:
  - A larger absolute value of $a$ makes the parabola narrower.
  - A smaller absolute value of $a$ makes the parabola wider.

#### 2. **$x$: The Variable**
- $x$ is the variable (or unknown) in the equation.
- The goal of solving a quadratic equation is often to find the values of $x$ that satisfy the equation (i.e., the roots or solutions).
- $x$ can take on any real number value, depending on the context of the problem.

#### 3. **$b$: The Coefficient of $x$**
- $b$ is the coefficient of the linear term ($x$).
- It influences the slope of the parabola and helps determine the position of the vertex (the highest or lowest point of the parabola).
- If $b = 0$, the parabola is symmetric about the $y$-axis.

#### 4. **$c$: The Constant Term**
- $c$ is the constant term in the equation.
- It represents the $y$-intercept of the parabola, which is the point where the graph crosses the $y$-axis.
- When $x = 0$, the value of $y$ is equal to $c$.

---

### Example:
Consider the quadratic equation:

$$
2x^2 - 4x + 1 = 0
$$

Here:
- $a = 2$: The parabola opens upwards because $a > 0$. The parabola is relatively narrow because $|a| = 2$ is not very small.
- $b = -4$: This affects the slope of the parabola and shifts its position.
- $c = 1$: The parabola crosses the $y$-axis at the point $(0, 1)$.

---

### Key Points About Each Component:
- **$a$:** Controls the shape and direction of the parabola.
- **$x$:** The variable whose values we solve for.
- **$b$:** Influences the slope and position of the parabola.
- **$c$:** Determines the $y$-intercept of the parabola.

Understanding these components helps you analyze and solve quadratic equations effectively!

# Regression Metrics: Understanding MSE, MAE, RMSE and MAD


When you build a model to predict numbers‚Äîlike house prices, temperatures, or sales‚Äîhow do you know if it‚Äôs doing a good job? That‚Äôs where _regression metrics_ come in. They‚Äôre like scorecards that tell you how close your predictions are to the real values. Let‚Äôs explore three popular ones: **Mean Absolute Error (MAE)**, **Mean Squared Error (MSE)**, and **Mean Absolute Deviation (MAD)**, and see how they work in everyday situations.

## 1. Mean Squared Error (MSE)

### What is MSE?
MSE measures the average of the squared differences between predicted and actual values. By squaring the errors, it penalizes larger discrepancies more heavily than smaller ones.

### Formula:
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$

- $ y_i $: Actual value  
- $ \hat{y}_i $: Predicted value  
- $ n $: Number of data points  

### Key Characteristics:
- **Sensitivity to Outliers**: Because MSE squares the differences, it gives more weight to large errors. This makes it ideal for scenarios where outliers are important.
- **Units**: The result is in squared units of the data, which can make interpretation less intuitive.

### Real-World Scenario: Stock Price Prediction
Imagine you're building a model to predict stock prices. Even small errors in stock price predictions can lead to significant financial losses. Since MSE heavily penalizes large prediction errors, it ensures the model focuses on minimizing these deviations. For example, if your model predicts a stock price of $100 but the actual price is $120, the squared error will be much larger than if the actual price was $105. This makes MSE a good choice for scenarios where large errors are undesirable.

---

### Simple Example of Mean Squared Error (MSE)

Let‚Äôs calculate the **Mean Squared Error (MSE)** step by step with a small example.

---

### Scenario:
You have a model that predicts house prices, and you want to evaluate how well it performs. Here are the actual prices ($y$) and the predicted prices ($\hat{y}$):

| Actual Price ($y$) | Predicted Price ($\hat{y}$) | Difference ($y - \hat{y}$) | Squared Difference ($(y - \hat{y})^2$) |
|---------------------|-------------------------------|-----------------------------|-----------------------------------------|
| 100                 | 110                           | $100 - 110 = -10$          | $(-10)^2 = 100$                        |
| 150                 | 145                           | $150 - 145 = 5$            | $5^2 = 25$                             |
| 200                 | 190                           | $200 - 190 = 10$           | $10^2 = 100$                           |

---

### Step 1: Calculate the Squared Differences
For each data point, compute the squared difference between the actual value ($y$) and the predicted value ($\hat{y}$):
- For the first house: $(100 - 110)^2 = (-10)^2 = 100$
- For the second house: $(150 - 145)^2 = 5^2 = 25$
- For the third house: $(200 - 190)^2 = 10^2 = 100$

---

### Step 2: Compute the Average
Now, take the average of these squared differences:
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$
Here:
- $n = 3$ (number of data points),
- The squared differences are $100, 25, 100$.

$$
\text{MSE} = \frac{100 + 25 + 100}{3} = \frac{225}{3} = 75
$$

---

### Final Result:
The **Mean Squared Error (MSE)** for this example is **75**.

---

### Interpretation:
- The MSE measures the average squared difference between the actual and predicted values.
- A lower MSE indicates better performance of the model. In this case, an MSE of 75 suggests that, on average, the model's predictions are off by about $\sqrt{75} \approx 8.66$ units (if you take the square root to interpret it in the original units).

## 2. Mean Absolute Error (MAE)

### What is MAE?
MAE measures the average of the absolute differences between predicted and actual values. Unlike MSE, it treats all errors equally, regardless of their size. It's a way to measure how "wrong" predictions are, on average, compared to the actual values.

### Formula:
$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i|
$$

- $ y_i $: Actual value  
- $ \hat{y}_i $: Predicted value  
- $ n $: Number of data points  

### Key Characteristics:
- **Robustness to Outliers**: MAE is less sensitive to outliers because it doesn't square the errors. This makes it a better choice when outliers are not critical to the analysis.
- **Units**: The result is in the same units as the data, making it easier to interpret.

### Simple Example of Mean Absolute Error (MAE)

Let‚Äôs calculate the **Mean Absolute Error (MAE)** step by step with a small example.

### Scenario:
You have a model that predicts house prices, and you want to evaluate how well it performs. Here are the actual prices ($y$) and the predicted prices ($\hat{y}$):

| Actual Price ($y$) | Predicted Price ($\hat{y}$) | Difference ($y - \hat{y}$) | Absolute Difference ($\|y - \hat{y}\|$) |
|---------------------|------------------------------|-----------------------------|---------------------------------------|
| 100                 | 110                          | $100 - 110 = -10$          | $\|-10\| = 10$                           |
| 150                 | 145                          | $150 - 145 = 5$            | $\|5\| = 5$                              |
| 200                 | 190                          | $200 - 190 = 10$           | $\|10\| = 10$                            |

### Step 1: Calculate the Absolute Differences
For each data point, compute the absolute difference between the actual value ($y$) and the predicted value ($\hat{y}$):
- For the first house: $|100 - 110| = |-10| = 10$
- For the second house: $|150 - 145| = |5| = 5$
- For the third house: $|200 - 190| = |10| = 10$

### Step 2: Compute the Average
Now, take the average of these absolute differences:
$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i|
$$
Here:
- $n = 3$ (number of data points),
- The absolute differences are $10, 5, 10$.

$$
\text{MAE} = \frac{10 + 5 + 10}{3} = \frac{25}{3} \approx 8.33
$$

---

### Final Result:
The **Mean Absolute Error (MAE)** for this example is approximately **8.33**.

---

### Interpretation:
- The MAE measures the average absolute difference between the actual and predicted values.
- A lower MAE indicates better performance of the model. In this case, an MAE of approximately 8.33 suggests that, on average, the model's predictions are off by about **8.33 units** in the original scale.

---

### Why Use MAE?
- Unlike MSE, MAE treats all errors equally, regardless of their size. This makes it less sensitive to outliers and more interpretable because the result is in the same units as the data.

## 3. Mean absolute deviation (MAD) or L1 Norm
The mean absolute deviation of a dataset is the average distance between each data point and the mean. It gives us an idea about the variability in a dataset.

$$\text{MAD}=\dfrac{\sum{\lvert x_i-\bar{x} \rvert}}{n}$$

Here's how to calculate the mean absolute deviation.

- Find the Average (Mean):
First, add up all the numbers in your list and then divide by how many numbers there are. This gives you the average or "mean" value.<br><br>
- Calculate Deviations:
For each number in your list, figure out how far it is from the average. If the number is bigger than the average, subtract the average from it. If it's smaller, subtract it from the average. This difference is the "deviation."<br><br>
- Make Deviations Positive:
Since you want to measure how far numbers are from the mean regardless of whether they're above or below it, you take the absolute value of each deviation. That means you just look at the size of the number, ignoring if it's positive or negative.<br><br>
- Average These Deviations:
Now, add up all these absolute deviations and divide by the number of items in your list. This gives you the Mean Absolute Deviation.<br><br>
Example:<br><br>

Suppose you have the numbers: $2, 4, 6, 8$.<br>
Mean: $(2 + 4 + 6 + 8) / 4 = 5$<br>
Deviations:<br>

$2 - 5 = -3 (absolute value = 3)$<br>
$4 - 5 = -1 (absolute value = 1)$<br>
$6 - 5 = 1$<br>
$8 - 5 = 3$<br>
Sum of Absolute Deviations: $3 + 1 + 1 + 3 = 8$<br>
MAD: $8 / 4 = 2$

- $\bar{x}$ is the mean of all the $x_i$ values.
- $x_i$ represents each individual data point in your sample or population.
- $|x_i - \bar{x}|$ is the absolute deviation of each $x_i$ from the mean $\bar{x}$.
- $\sum$ means sum up all these absolute deviations.
- $n$ is the number of observations or data points.

Following these steps in the example below is probably the best way to learn about mean absolute deviation. <br>
Let's find the mean absolute deviation.

````python
sample = np.array([10, 15, 15, 17, 18, 21])
mean = np.mean(sample)
print(f"MEAN: {mean}")

deviation = np.absolute([x for x in sample - mean])
print(f"DEVIATION: {deviation}")

deviations_avg = np.mean(deviation)
print(f"DEVIATION AVERAGE: {deviations_avg}")
````

## 4. Root Mean Squared Error "RMSE" or L2 Norm
 It is simply the square root of the Mean Squared Error (MSE) . While MSE gives you the average of the squared differences between predicted and actual values, RMSE converts this back into the original units of the data, making it easier to interpret.
 - RMSE provides a measure of error in the same units as the data (e.g., dollars, meters, etc.), which makes it more intuitive than MSE.
- Like MSE, RMSE penalizes larger errors more heavily because it‚Äôs derived from squared differences.

#### Why Use RMSE?
- RMSE provides a measure of error in the same units as the data (e.g., dollars, meters, etc.), which makes it more intuitive than MSE.
- Like MSE, RMSE penalizes larger errors more heavily because it‚Äôs derived from squared differences.

---

### Example of RMSE

Let‚Äôs continue with the example from the table:

| Actual ($y$) | Predicted ($\hat{y}$) | Difference | Squared Difference | Absolute Difference |
|---------------|-------------------------|------------|---------------------|----------------------|
| 10            | 12                      | -2         | 4                   | 2                    |
| 15            | 20                      | -5         | 25                  | 5                    |
| 20            | 18                      | 2          | 4                   | 2                    |

We already calculated:
- **MSE**: $ \text{MSE} = \frac{1}{3} \times (4 + 25 + 4) = 11 $

Now, to calculate **RMSE**, take the square root of the MSE:
$$
\text{RMSE} = \sqrt{\text{MSE}} = \sqrt{11} \approx 3.32
$$

So, the **RMSE** for this example is approximately **3.32**.

---

### Interpretation of RMSE
- The RMSE value of **3.32** means that, on average, the model's predictions are off by about **3.32 units** in the original scale (e.g., if predicting house prices, the error would be in dollars; if predicting temperature, the error would be in degrees).
- Unlike MSE, RMSE is in the same units as the data, making it easier to understand and compare.

## Conclusion

Choosing the right regression metric depends on the specific requirements of your problem and the nature of your data. If large errors are particularly problematic, **MSE** is the way to go. For a straightforward, interpretable measure of average error, **MAE** is ideal. And when dealing with datasets that contain significant outliers, **MAD** provides a robust alternative. By understanding these metrics and their applications, you can effectively evaluate and improve the performance of your regression models in real-world scenarios.

## Derivates


A derivative measures how something changes at a specific moment. It tells you the rate of change of one quantity (like position, temperature, or profit) with respect to another (like time or distance). This is the central idea of Differential Calculus, known as derivatives.

Let‚Äôs say you‚Äôre driving a car, and you want to know how fast you‚Äôre going at any moment. That‚Äôs what differential calculus is all about ‚Äî figuring out how things change.

If you look at your speedometer, it tells you your speed right now. That‚Äôs the rate of change of your position. In math terms, this "rate of change" is called a derivative .
So, if you‚Äôre driving along a road, the derivative tells you:
"How much your position changes as time passes."

### How Does It Work?
Let‚Äôs say you have a graph that shows how far you‚Äôve driven over time: <br><br>

On the graph:<br>
The horizontal axis (x-axis) is time.<br>
The vertical axis (y-axis) is distance.<br>
If you draw a line between two points on the graph, the slope of that line tells you your average speed over that time period.<br> But if you zoom in closer and closer to just one point, the slope of the line at that tiny spot tells you your instantaneous speed ‚Äî that‚Äôs the derivative!

Let‚Äôs create a simple graph in Python to visualize the concept of derivatives. We‚Äôll use a basic example: a car driving along a straight road, where its position changes over time.

We‚Äôll plot:

The car‚Äôs position over time (a curve).
The derivative of the position, which represents the car‚Äôs speed at any given moment.

````python
# Define the position function (e.g., position over time)
# Let's say the car's position is given by: position(t) = t^2 (a parabola)
def position(t):
    return t**2

# Define the derivative (speed) of the position function
# The derivative of t^2 is 2*t
def speed(t):
    return 2 * t

# Generate time values from 0 to 5 seconds
time = np.linspace(0, 5, 100)  # 100 points between 0 and 5

# Calculate position and speed for each time value
positions = position(time)
speeds = speed(time)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the position vs time
plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st subplot
plt.plot(time, positions, label="Position (t^2)", color="blue")
plt.title("Position Over Time")
plt.xlabel("Time (seconds)")
plt.ylabel("Position (meters)")
plt.grid(True)
plt.legend()


# Plot the speed vs time (derivative of position)
plt.subplot(2, 1, 2)  # 2 rows, 1 column, 2nd subplot
plt.plot(time, speeds, label="Speed (2*t)", color="red")
plt.title("Speed Over Time (Derivative of Position)")
plt.xlabel("Time (seconds)")
plt.ylabel("Speed (meters/second)")
plt.grid(True)
plt.legend()

plt.
# Show the plots
plt.tight_layout()
plt.show()
````

### What Are Derivative Rules?

Derivative rules are formulas that allow us to compute the derivatives of functions systematically. Instead of calculating derivatives from scratch every time, we use these rules to simplify the process. These rules apply to different types of functions (e.g., polynomials, trigonometric functions, exponential functions) and combinations of functions (e.g., sums, products, compositions).

Below is a table summarizing the most common derivative rules:

---

### Table of Common Derivative Rules

| **Rule Name**         | **Function $ f(x) $**                     | **Derivative $ f'(x) $**                      |
|------------------------|--------------------------------------------|-----------------------------------------------|
| **Constant Rule**      | $ f(x) = c $                              | $ f'(x) = 0 $                                |
| **Power Rule**         | $ f(x) = x^n $                            | $ f'(x) = n \cdot x^{n-1} $                 |
| **Sum/Difference Rule**| $ f(x) = g(x) \pm h(x) $                  | $ f'(x) = g'(x) \pm h'(x) $                 |
| **Constant Multiple**  | $ f(x) = c \cdot g(x) $                   | $ f'(x) = c \cdot g'(x) $                   |
| **Product Rule**       | $ f(x) = g(x) \cdot h(x) $                | $ f'(x) = g'(x) \cdot h(x) + g(x) \cdot h'(x) $ |
| **Quotient Rule**      | $ f(x) = \frac{g(x)}{h(x)} $              | $ f'(x) = \frac{g'(x) \cdot h(x) - g(x) \cdot h'(x)}{(h(x))^2} $ |
| **Chain Rule**         | $ f(x) = g(h(x)) $                        | $ f'(x) = g'(h(x)) \cdot h'(x) $            |
| **Exponential Rule**   | $ f(x) = e^x $                            | $ f'(x) = e^x $                             |
| **Logarithmic Rule**   | $ f(x) = \ln(x) $                         | $ f'(x) = \frac{1}{x} $                     |

## What is a Scenet Line?

A secant line is a straight line that passes through two points on a curve.

````python
# Define the function h(x)
def h(x):
    return np.sqrt(15 - 2 * x)

# Calculate the y-values for x = 3 and x = 7
x1, x2 = 3, 7
y1, y2 = h(x1), h(x2)

# Define the slope of the secant line
slope = (y2 - y1) / (x2 - x1)

# Define the equation of the secant line: y = slope * (x - x1) + y1
def secant_line(x):
    return slope * (x - x1) + y1

# Generate x values for plotting the function
x_vals = np.linspace(0, 7.8, 500)  # x values from 0 to 7
y_vals = h(x_vals)              # y values for the function h(x)

# Generate x values for plotting the secant line
secant_x_vals = np.linspace(0, 8, 500)  # Extend the secant line beyond the points
secant_y_vals = secant_line(secant_x_vals)  # y values for the secant line

# Plot the function and the secant line
plt.figure(figsize=(10, 6))

# Plot the function h(x)
plt.plot(x_vals, y_vals, label="h(x) = ‚àö(15 - 2x)", color="blue")

# Plot the secant line
plt.plot(secant_x_vals, secant_y_vals, label="Secant Line", color="red", linestyle="--")

# Plot the two points (x1, y1) and (x2, y2)
plt.scatter([x1, x2], [y1, y2], color="black", zorder=5)  # Highlight the points
plt.text(x1, y1, f"({x1}, {y1:.2f})", fontsize=10, verticalalignment="bottom")
plt.text(x2, y2, f"({x2}, {y2:.2f})", fontsize=10, verticalalignment="bottom")

# Add labels, legend, and grid
plt.title("Secant Line Demonstration")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")  # x-axis
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")  # y-axis
plt.grid(True)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
````

## What is a Tangent Line?


A tangent line is a straight line that just "touches" a curve at a single point without crossing it. At that point, the tangent line has the same slope as the curve. The slope of the tangent line represents the instantaneous rate of change of the function at that point.

In calculus terms:

The slope of the tangent line is given by the derivative of the function at that point.

The tangent line provides a way to approximate complex, nonlinear behavior with a simple straight line. This makes it easier to:

Analyze rates of change.
Predict future behavior.
Optimize processes.
Solve problems in science, engineering, and beyond.
In short, the tangent line is a tool that helps us understand and interact with the world in a precise and meaningful way!

````python
# Define the function
def f(x):
    return x**2  # Example: f(x) = x^2

# Derivative of the function (slope of the tangent line)
def f_prime(x):
    return 2 * x  # Derivative of f(x) = x^2 is f'(x) = 2x

# Point where we want the tangent line
x_point = 1  # Tangent at x = 1
y_point = f(x_point)  # Corresponding y-value

# Slope of the tangent line at x = 1
slope = f_prime(x_point)

# Equation of the tangent line: y = slope * (x - x_point) + y_point
def tangent_line(x):
    return slope * (x - x_point) + y_point

# Generate x values for plotting
x_vals = np.linspace(-2, 2, 500)  # Range of x-values for the curve
y_vals = f(x_vals)               # y-values for the curve

# Generate x values for the tangent line
tangent_x_vals = np.linspace(x_point - 1, x_point + 1, 100)  # Narrow range around x_point
tangent_y_vals = tangent_line(tangent_x_vals)               # y-values for the tangent line

# Plot the curve and the tangent line
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label="f(x) = x^2", color="blue")  # Original curve
plt.plot(tangent_x_vals, tangent_y_vals, label="Tangent Line", color="red", linestyle="--")  # Tangent line
plt.scatter(x_point, y_point, color="black", zorder=5)  # Highlight the point of tangency
plt.text(x_point, y_point, f"({x_point}, {y_point})", fontsize=10, verticalalignment="bottom")

# Add labels, legend, and grid
plt.title("Tangent Line Visualization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")  # x-axis
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")  # y-axis
plt.grid(True)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
````

## Gradient

The gradient is a mathematical tool that tells you how something changes in different directions. It‚Äôs like a "compass" that points in the direction of the steepest increase of a function, and it also tells you how fast things are changing in that direction.

If you‚Äôre working with a function that depends on multiple variables (like height depending on both x and y), the gradient gives you a vector (a fancy word for an arrow) that summarizes all the rates of change at once.

The terms gradient and derivative are closely related, but they are used in different contexts and have distinct meanings. Let‚Äôs break it down step by step to clarify the differences.

The derivative measures how a function changes as its input changes. It‚Äôs a fundamental concept in calculus and is typically used for functions of one variable .

The gradient generalizes the idea of a derivative to functions of multiple variables . Instead of just one rate of change, the gradient gives you a vector that summarizes all the rates of change in every direction.

### Gradient in Math Terms

Let‚Äôs make this more precise:

#### For a Single Variable Function:
If you have a function like $ f(x) $, the rate of change is just the derivative $ f'(x) $. This is like the slope of a tangent line.

#### For a Multivariable Function:
If you have a function like $ f(x, y) $ (which depends on two variables), the **gradient** is a vector that combines all the partial derivatives (rates of change in each direction). It‚Äôs written as:

$$
\nabla f = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
$$

- $ \frac{\partial f}{\partial x} $: How $ f $ changes if you move in the $ x $-direction.
- $ \frac{\partial f}{\partial y} $: How $ f $ changes if you move in the $ y $-direction.

The gradient vector $ \nabla f $ points in the direction where $ f $ increases the fastest.

````python
def f(x, y):
  return x**2 + y**2

# Define the gradient of the function
def gradient_f(x, y):
  df_dx = 2 * x
  df_dy = 2 * y
  return np.array([df_dx, df_dy])

# Point at which to calculate the gradient
point = np.array([3, 4])

# Calculate the gradient at the given point
grad = gradient_f(point[0], point[1])

# Create a meshgrid for plotting the function
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the function
plt.figure(figsize=(10, 6))
plt.contour(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(label='f(x, y)')

# Plot the gradient vector
plt.quiver(point[0], point[1], grad[0], grad[1], color='red', scale=10, label='Gradient (6, 8)')

# Highlight the point
plt.scatter(point[0], point[1], color='red')
plt.text(point[0], point[1], f'({point[0]}, {point[1]})', fontsize=12, verticalalignment='bottom')

# Add labels and legend
plt.title('Gradient of f(x, y) = x^2 + y^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
````
